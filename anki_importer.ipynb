```python
import genanki
import csv
import os
```

```python
my_deck = genanki.Deck(
  2059400111,
  'Swedish Wordlist Sound Only')
```

```python
my_model = genanki.Model(
  1091735104,
  'Simple Model with Media',
  fields=[
    {'name': 'Swedish'},
    {'name': 'English'},
    {'name': 'Sound'},                                  # ADD THIS
  ],
  templates=[
    {
      'name': 'Card 1',
      'qfmt': '<p style="text-align: center;font-size: 31px;"> <b>{{Swedish}}</b> {{Sound}}',              # AND THIS
      'afmt': '{{FrontSide}}<hr id="answer">{{English}}',
    },
  ])
```

```python
with open('wordlist.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        my_note = genanki.Note(
            model=my_model,
            fields=[row['se'], row['en'],f'[sound:{row["id"]}.mp3]'])
        my_deck.add_note(my_note)
```

```python
my_package = genanki.Package(my_deck)
```

```python
my_package.media_files = [f"titty_sounds/{i}" for i in os.listdir(path='titty_sounds')]
```

```python
my_package.write_to_file('swedish_wordlist_sound_only.apkg')
```
